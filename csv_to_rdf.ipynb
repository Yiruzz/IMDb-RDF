```python
sparql_query = """
PREFIX ex: <http://example.org/movies#>

CONSTRUCT {
  ?movie a ex:Movie ;
         ex:title ?title ;
         ex:releaseYear ?releaseYear ;
         ex:certificate ?certificate ;
         ex:runtime ?runtime ;
         ex:genre ?genre ;
         ex:imdbRating ?imdbRating ;
         ex:metaScore ?metaScore ;
         ex:director ?director ;
         ex:voteCount ?voteCount ;
         ex:gross ?gross ;
         ex:stars ?stars .
}
WHERE {
  BIND(IRI(CONCAT("http://example.org/movies#", REPLACE(?Series_Title, " ", "_"))) AS ?movie)
  BIND(?Series_Title AS ?title)
  BIND(?Released_Year AS ?releaseYear)
  BIND(?Certificate AS ?certificate)
  BIND(?Runtime AS ?runtime)
  BIND(?Genre AS ?genre)
  BIND(?IMDB_Rating AS ?imdbRating)
  BIND(?Meta_score AS ?metaScore)
  BIND(?Director AS ?director)
  BIND(?No_of_Votes AS ?voteCount)
  BIND(?Gross AS ?gross)
  BIND(?Stars AS ?stars)
}
"""

# Guardamos el archivo SPARQL con el mapeo completo en el entorno de trabajo
with open("movies.sparql", "w") as file:
    file.write(sparql_query)
```

## Enriquecimiento de Datos

Script de ejemplo de como encontrar un nodo y sacarle valores utiles a wikidata:

```python
import requests

data = ["capital_cost", "IMDb ID", "narrative_location", "production_company"]
seq = ["follows", "followed by"]
def search_movie_on_wikidata(title, year):
    # Define la consulta SPARQL con el título y el año de la película
    #SELECT ?movie ?movieLabel ?imdb_id ?directorLabel ?actorLabel WHERE {
    query = """
    SELECT DISTINCT ?movie ?imdb_id ?capital_cost ?narrative_location ?production_company WHERE {
    ?movie wdt:P31 wd:Q11424 ;
            wdt:P1476 ?title ;
            wdt:P577 ?release_date ;
            rdfs:label ?label ;
            wdt:P345 ?imdb_id ;
            wdt:P2130 ?capital_cost ;
            wdt:P840 ?narrative_location ;
            wdt:P272 ?production_company .


    FILTER((lang(?label) = "en") && (lcase(str(?title)) = "%s"))
    FILTER(YEAR(?release_date) = %d)
    }
    """ % (title.lower(), year)

    url = "https://query.wikidata.org/sparql"
    response = requests.get(url, params={'query': query, 'format': 'json'})

    # Verifica si la respuesta fue exitosa
    if response.status_code == 200:
        data = response.json()

        for result in data['results']['bindings']:
            print(result)
    else:
        print(f"Error: {response.status_code}")
        print(response.text)

search_movie_on_wikidata("The Shawshank Redemption", 1994)
```
